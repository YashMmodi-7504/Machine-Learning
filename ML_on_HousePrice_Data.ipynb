```python
## Prepare data for Machine Learning
# We know that the machine learning algorithms are mathematical expressions, so we need to provide the data to the algorithm in specific format
# Encoding
# Step-1 : Divide the data based on their datatypes
df_cat = df_houseprice.select_dtypes(include=['object'])
df_cat.columns
```

```python
# Use Dummy Encoding
dummy_var =pd.get_dummies(data = df_cat, drop_first = True)
dummy_var = dummy_var.reset_index()
```

```python
# Use Dummy Encoding
dummy_var2 =pd.get_dummies(data = df_cat)
dummy_var2
```

```python
df_num = df_houseprice.select_dtypes(include = [np.number])
df_num.head()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
df_target = df_num['Price']
df_num = df_num.drop('Price', axis = 1)
```

```python
# Initiate the StandardScaler
ss = StandardScaler()
```

```python
# Transform the Data
Scale = ss.fit_transform(df_num)
```

```python
Scale
```

```python
df_Scale = pd.DataFrame(Scale, columns = df_num.columns)
df_Scale = df_Scale.reset_index()
```

```python
df_Scale.head()
```

```python
# Make a DataFrame for independent variables
df_independent = pd.concat([df_Scale, dummy_var], axis = 1)
df_independent.head()
```

```python
import statsmodels
import statsmodels.api as sm
df_independent = sm.add_constant(df_independent)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(df_independent,
                                                   df_target,
                                                  random_state = 10,
                                                  test_size = 0.2)
```

```python
print('X_train', X_train.shape)
print('Y_train', Y_train.shape)

print('X_test', X_test.shape)
print('Y_test', Y_test.shape)
```

# Linear Regression

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
model = LinearRegression()
model.fit(X_train, Y_train)
predictions = model.predict(X_test)
mse = mean_squared_error(Y_test, predictions)
print("Mean Squared Error:",mse)
```

```python
predictions
```

```python
from sklearn.metrics import mean_absolute_error
mae = mean_absolute_error(Y_test, predictions)
print("Mean Absolute Error:",mae)
```

```python
rms = mse**0.5
print("root_mean_square",rms)
```
